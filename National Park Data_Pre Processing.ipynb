```python
import pandas as pd

df1 = pd.read_excel('Query Builder for Historic Annual Recreation Visits (1904 - 1979).xlsx')
#data location: https://irma.nps.gov/Stats/SSRSReports/National%20Reports/Query%20Builder%20for%20Historic%20Annual%20Recreation%20Visits%20(1904%20-%201979)

df2 = pd.read_excel('Query Builder for Public Use Statistics (1979 - Last Calendar Year).xlsx') 
#data location: https://irma.nps.gov/Stats/SSRSReports/National%20Reports/Query%20Builder%20for%20Public%20Use%20Statistics%20(1979%20-%20Last%20Calendar%20Year)

df3 = pd.read_excel('US National Parks Metadata.xlsx')
#data location: https://en.wikipedia.org/wiki/List_of_national_parks_of_the_United_States
```

```python
#df3 (metadata dataframe)

df3[['State', 'Coordinates']] = df3['Location'].str.split('\n', expand=True)
df3[['Latitude', 'Longitude']] = df3['Coordinates'].str.split(' ', expand=True)

#drop the intermediate columns
df3.drop(['Location', 'Coordinates'], axis=1, inplace=True)
df3.drop('Park', axis=1, inplace=True)

# print(df3)
df3.head()

```

```python
#df1:

df1.rename(columns={'Total Recreation\nVisits':'Recreation Visits'}, inplace=True)

df1['Park'] = df1['Park'].str.replace(' NP$', '', regex=True) 
df1['Park'] = df1['Park'].str.replace('NP & PRES','', regex=True)
df1['Park'] = df1['Park'].str.replace(' NP ','', regex=True)
df1['Month'] = ''
df1['Historical'] = 'Y'
# df1.drop('State',axis=1, inplace=True)
# df1.head()

#enrich df1 with metadata

df_historical = pd.merge(df1,df3,on='Unit Code')
# df_historical.drop('State_x',axis=1, inplace=True)
df_historical = df_historical.reindex(columns=['Park', 'Unit Code', 'Region', 'Year', 'Month', 'Recreation Visits', 'Historical', 'Date Established', 'Area in acres (2023)', 'Description', 'State', 'Latitude', 'Longitude'])

df_historical.head()
```

```python
#df2:
df2['Park'] = df2['Park'].str.replace(' NP$', '', regex=True) 
df2['Park'] = df2['Park'].str.replace('& PRES','', regex=True)
df2['Park'] = df2['Park'].str.replace(' NP ','', regex=True)
#create column for historical (Y/N)
df2['Historical'] = 'N'
# df2.drop('State',axis=1, inplace=True)
df2.head()

#enrich df2 with metadata

df_current = pd.merge(df2,df3,on='Unit Code')
# df_current.drop('State_x', axis=1, inplace=True)

df_current = df_current.reindex(columns=['Park', 'Unit Code', 'Region', 'Year', 'Month', 'Recreation Visits', 'Historical', 'Date Established', 'Area in acres (2023)', 'Description', 'State', 'Latitude', 'Longitude'])

df_current.head()
```

```python
#union df 1 and df 2
final_df = pd.concat([df_historical,df_current],ignore_index=True)

final_df.head()
```

```python
final_df.to_excel('US National Park Visitation.xlsx', index=False)
```
